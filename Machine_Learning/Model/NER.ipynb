```
import pandas as pd
import numpy as np
df1 = pd.read_csv("df1.csv")
df2 = pd.read_csv("df2.csv")
```

```
def fun_ner(text):
  l=""
  x = ner_model(text)
  for j in range(len(x[1][0])):
    if(x[1][0][j] == 'B-WORK_OF_ART' or x[1][0][j] == 'I-WORK_OF_ART' or x[1][0][j] == 'B-PRODUCT' or x[1][0][j] == 'I-PRODUCT'):  
      l = " " + x[0][0][j]
  return l
```

```
df1["Product"] = ""
df2["Product"] = ""
```

```

for i in df1.index:
  text = df1.at[i,"Text"]
  try:
    ner = fun_ner([text])
    df1.at[i,'Product'] = ner
  except:
    df1.at[i,'Product'] = ""
  
  




  
```
